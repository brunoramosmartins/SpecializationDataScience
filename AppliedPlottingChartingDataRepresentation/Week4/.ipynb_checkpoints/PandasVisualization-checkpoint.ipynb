### DataFrame.plot

Agora, onde a visualização integrada de pandas realmente brilha é ajudar na plotagem rápida e fácil de séries e DataFrames que podem nos ajudar a explorar os dados. Vamos fazer um DataFrame. Primeiro, definiremos a semente para o gerador de números aleatórios, que nos permitirá reproduzir os dados.

Em seguida, vamos adicionar três colunas de dados aleatórios de série temporal.

Podemos gerar os dados aleatórios somando cumulativamente números aleatórios.

numpy tem uma ótima função para isso chamada [`cumsum`](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), que soma cumulativamente um array

Vamos fazer isso para três colunas, A, B e C. E também deslocar a coluna B em + 20 e a coluna C em - 20. Usando [date_range](https://pandas.pydata.org/docs/reference/api/pandas.date_range.html), podemos definir o índice para ser diário em 2017.

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0), 
                   'B': np.random.randn(365).cumsum(0) + 20,
                   'C': np.random.randn(365).cumsum(0) - 20}, 
                  index=pd.date_range('1/1/2017', periods=365))
df.head()
```

Agora vamos ver como esses dados se parecem um pouco mais visualmente para plotar esses dados, podemos usar o método plot no DataFrame.

```python
df.plot(); # add a semi-colon to the end of the plotting call to suppress unwanted output
```

O método plot em series e DataFrame é apenas um wrapper simples em torno de plt.plot. Portanto, quando chamamos df.plot, obtemos um gráfico de linha de todas as colunas no quadro de dados com rótulos.

Observe como as cores são ligeiramente diferentes das cores padrão do matplotlib por causa do estilo que usamos.

Além disso, observe este truque legal do Jupyter Notebook. Se você adicionar um ponto e vírgula ao final da chamada de plotagem, isso suprimirá a saída indesejada. E isso também funciona com matplotlib regular.

DataFrame.plot nos permite traçar vários tipos diferentes de plotagens. Podemos selecionar qual gráfico queremos usar pressionando-o no parâmetro kind.

Vamos tentar criar um gráfico de dispersão que leva as colunas A e B de df.

Passamos em scatter para o parâmetro kind para alterar o tipo de plotagem.

Isso é o que obteríamos se chamássemos plt.plot nas colunas A e B com o parâmetro marcador de ponto.

Passar dispersão para o argumento de palavra-chave gentil mudou o gráfico para um gráfico de dispersão. Você também pode escolher o tipo de gráfico usando os métodos dataframe.plot.kind em vez de fornecer o argumento de palavra-chave kind.

```python
df.plot('A','B', kind = 'scatter');
```

Você também pode escolher o tipo de gráfico usando os métodos `DataFrame.plot.kind` em vez de fornecer o argumento de palavra-chave `kind`.

`tipo`:
- `'line'`: gráfico de linha (padrão)
- `'bar'`: gráfico de barra vertical
- `'barh'`: gráfico de barra horizontal
- `'hist'`: histograma
- `'caixa'`: boxplot
- `'kde'`: gráfico de estimativa de densidade do kernel
- `'densidade'`: o mesmo que 'kde'
- `'área'`: plotagem da área
- `'torta'`: gráfico de pizza
- `'scatter'`: gráfico de dispersão
- `'hexbin'`: gráfico hexbin

Agora vamos tentar isso e criar algo um pouco mais complexo. Desta vez, queremos fazer um gráfico de dispersão com pontos variando em cor e tamanho. Usaremos df.plot.scatter, passaremos nas colunas A e C.

E defina a cor C e o tamanho S para alterar com base no valor da coluna B.

Por fim, podemos escolher a paleta de cores usada passando uma string para o mapa de cores do parâmetro.

Aqui, usarei viridis, que é particularmente agradável aos olhos.

```python
# create a scatter plot of columns 'A' and 'C', with changing color (c) and size (s) based on column 'B'
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis');
```

Aqui podemos ver as colunas A e C plotadas uma contra a outra com o tamanho e a cor mudando com base nos valores da coluna B. Como df.plot.scatter retorna um matplotlib.axes._subplot, podemos realizar modificações nesse objeto da mesma forma que os objetos retornados pelos gráficos de matplotlib.

Por exemplo, vamos ver como fica quando alteramos a proporção de AxesSubplot para igual.

Definir a proporção como igual permite que o visualizador veja facilmente que o intervalo da série A é muito menor do que a série C.

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')
```

Usando pandas, também podemos facilmente fazer gráficos de caixa, histogramas e gráficos de estimativa de densidade de kernel. Que são úteis para visualizar uma estimativa da função de densidade de probabilidade de uma variável. Os gráficos de estimativa de densidade do kernel são úteis em aplicativos de ciência de dados em que você deseja derivar uma função contínua suave de uma determinada amostra.

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.7);
```

[Kernel density estimation plots](https://en.wikipedia.org/wiki/Kernel_density_estimation) são úteis para derivar uma função contínua suave de uma determinada amostra.

```python
df.plot.kde();
```

### pandas.tools.plotting

O Pandas também possui ferramentas de plotagem que ajudam a visualizar grandes quantidades de dados ou dados de alta dimensão.

Vamos explorar algumas dessas ferramentas carregando o conjunto de dados da flor de íris.

O conjunto de dados da íris é um conjunto de dados multivariado clássico, que inclui o comprimento da sépala, a largura da sépala, o comprimento da pétala e a largura da pétala para centenas de amostras de três espécies da flor da íris. pandas tem uma ferramenta de plotagem que nos permite criar uma matriz de dispersão a partir de um DataFrame.

[Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set)

```python
iris = pd.read_csv('assets/iris.csv')
iris.head()
```

Uma matriz de dispersão é uma maneira de comparar cada coluna em um DataFrame com todas as outras colunas de maneira pareada. Vamos usá-lo para visualizar o dataframe da íris e ver quais insights podemos obter de nossos dados.

A matriz de dispersão cria gráficos de dispersão entre as diferentes variáveis e histogramas ao longo das diagonais.

Isso nos permite ver rapidamente alguns dos padrões mais óbvios no conjunto de dados.

```python
pd.plotting.scatter_matrix(iris);
```

Olhando para os gráficos de dispersão, parece que existem alguns agrupamentos distintos dos pontos, o que pode indicar que pode haver algum agrupamento acontecendo. E embora isso pareça impressionante e certamente útil, perceba que nas últimas duas semanas você adquiriu todas as habilidades para construir manualmente esses tipos de plotagens, se desejar. Isso nada mais é do que uma grade quatro por quatro de subparcelas, com algumas plotagens histogramas e outras, scatterplots.

Vejamos mais uma ferramenta de plotagem em pandas que nos ajudará a visualizar dados multivariados. Pandas inclui uma ferramenta de plotagem para criar plotagens de coordenadas paralelas.

```python
plt.figure()
pd.plotting.parallel_coordinates(iris, 'Name');
```

Gráficos de coordenadas paralelas são uma maneira comum de visualizar dados multivariados de alta dimensão.

Cada variável no conjunto de dados corresponde a uma linha vertical paralela igualmente espaçada. Os valores de cada variável são então conectados por linhas entre cada observação individual.

Colorir as linhas por classe, neste exemplo a espécie de flor, permite ao observador ver mais facilmente qualquer padrão ou agrupamento.

Por exemplo, olhando para o nosso conjunto de dados de íris, podemos ver que o comprimento da pétala e a largura da pétala são duas variáveis que dividem as diferentes espécies de forma bastante clara. Com a íris virginica tendo as pétalas mais longas e largas. E a íris setosa possuindo as pétalas mais curtas e estreitas. Bem, é isso para visualização em pandas.
