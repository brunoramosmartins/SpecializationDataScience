Vamos começar este módulo com uma visão mais profunda das subtramas. Até agora, usamos um único objeto de eixo para traçar um único gráfico ou figura. Às vezes é útil mostrar dois enredos complexos lado a lado para o visualizador comparar. Matplotlib lida com isso com uma única figura. Vamos primeiro definir nosso back-end de renderização para o back-end do notebook e, em seguida, importar nosso módulo pyplot e NumPy, pois precisaremos de ambos.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot?
```

Se olharmos para a documentação do subplot, veremos que o primeiro argumento é o número de linhas, o segundo o número de colunas e o terceiro é o número do gráfico. No matplotlib, uma grade conceitual é sobreposta na figura. E um comando subplot permite que você crie eixos para diferentes partes desta grade. Por exemplo, se quisermos criar dois gráficos lado a lado, chamaríamos subplot com os parâmetros 1, 2 e 1. Isso nos permitiria use 1 linha, com 2 colunas, e defina o primeiro eixo como o eixo atual.

Então, vamos criar uma nova figura e depois uma nova subtrama com uma linha e duas colunas. O primeiro objeto de eixo em que o pyplot irá plotar é o lado esquerdo.

```python
# let's create a new figure
plt.figure()

# and make a subplot with 1 row, 2 columns, and current axis is 1st subplot axes
plt.subplot(1, 2, 1)

# now let's plot some fake linear data
linear_data = np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data, '-o');
```

Ok, ótimo, temos um gráfico fino no lado esquerdo. Se fizermos uma segunda chamada para subplot(), podemos indicar que também queremos plotar no lado direito.

```python
# Our previous data
plt.subplot(1, 2, 1)
linear_data = np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data, '-o')

# now our new data in the second subplot
plt.subplot(1, 2, 2)
exponential_data = linear_data**2 
plt.plot(exponential_data, '-o')
```

E isso é bom. Agora temos dois gráficos, cada um com seus próprios objetos de eixo. Agora, a norma com matplotlib é que você armazene o objeto de eixo que você recebe da subtrama. Mas você pode chamar a subtrama novamente. A qualquer momento com os parâmetros que lhe interessam para recuperar um determinado eixo.

Dê uma olhada nesta figura, porém, você notou algo estranho nesta imagem pensando em nossa primeira semana de curso? As duas imagens têm valores de eixos y diferentes. Isso seria um problema e poderia enganar o leitor se não encontrássemos uma maneira de travar o eixo entre dois gráficos. Ao criar uma nova subtrama, você pode compartilhar os eixos x, y ou ambos usando os parâmetros compartilhar x e compartilhar y.

Vamos limpar isso.

```python
# Let's plot the first figure and get a reference to that axes
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')
# now let's tell the next plot we want to share the y axes with the
# one from ax1 to ensure the two subplots share the same y axis
ax2 = plt.subplot(122, sharey=ax1)
plt.plot(exponential_data, '-x');
```

Aqui vamos nós. Dois gráficos lado a lado e bloqueamos o eixo y.

Agora, aqueles de vocês que estão prestando muita atenção vão notar que usei a função `subplot` uma segunda vez, mas não passei três parâmetros, apenas um. Os desenvolvedores do maplotlib permitem especificar a linha, as colunas e o número do gráfico que você deseja com três parâmetros ou um único parâmetro. Onde as centenas valorizam o primeiro argumento, as dezenas o segundo argumento e as unidades o terceiro argumento.

Um fato importante a ser lembrado é que a localização do gráfico na matriz de itens é o índice começando em um e não em zero, como seria a convenção se você estivesse usando algo como NumPy. Portanto, se você estiver iterando em uma matriz ou lista, crie subtramas, lembre-se de começar na posição mais um.

Agora, há uma boa função chamada subplots, observe o plural, que permite obter muitos objetos de eixo de uma só vez, e acho isso ótimo. Então, se quisermos obter uma grade de três por três com todos os intervalos dos eixos x e y bloqueados, podemos fazer assim.

```python
# create a 3x3 grid of subplots, using tuple unpacking
fig, ((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9)) = plt.subplots(3, 3, sharex=True, sharey=True)
# plot the linear_data on the 5th subplot axes 
ax5.plot(linear_data, '-');
```

A sintaxe parece um pouco boba, talvez porque estamos desempacotando os resultados da função subplots diretamente, mas é uma maneira eficaz de construir uma grade onde tudo compartilha um eixo. Os resultados, no entanto, parecem muito bons, mas observe que esse método desativa os rótulos y e x, exceto para os gráficos que estão no lado esquerdo ou na parte inferior da figura.

Claro, podemos apenas percorrer uma lista e traçar um gráfico de cada vez, não precisamos armazenar uma referência para cada eixo.

```python
# So, let's create a basic figure
fig = plt.gcf()
# Now let's iterate over 6 potential spots in our figure. I'm going to create a plot with 2 rows and 3 columns
for i in range(1, 7):
    # let's not plot something if we are at position 5 or 3, we'll leave these as holes just to see what happens
    if i != 5 and i != 3:
        # Now let's add a subplot, we specify the first two arguments as the structure we are expecting the 
        # figure to take, so 2 rows and 3 columns, and the third argument is the position of this item in the 
        # figure, in this case i. 
        # While we are iterating linearly this is mapped onto our 2x3 row/column space, so the
        # numbering goes from left to right and top to bottom
        ax = fig.add_subplot(2, 3, i)
        # And we'll just add some text to the figures to make it more clear which item went where.
        # Remember from the lecture on annotation that text goes at the bottom left corner by default, 
        # and we are using relative positioning
        ax.text(0.5, 0.2, str((i)),fontsize=18, ha='center')
```

## SPLOMS

Então, agora entendemos porque existem abstrações de Eixos em uma figura, porque uma figura pode ter vários
Axes objetos que mostram várias exibições de dados. Uma técnica comum de exploração visual da ciência de dados é chamada
o SPLOM, que significa matrizes de gráfico de dispersão. Estes são particularmente úteis para obter o relacionamento
entre um número de variáveis diferentes a partir de uma rápida olhada. Agora, um SPLOM é realmente semelhante ao que
Edward Tufte chamou de "pequeno múltiplo" -- um conjunto de visuais que analisam dados relacionados, mas aquela fatia que
dados em diferentes pequenos visuais para que você possa ver as árvores e a floresta ao mesmo tempo.

Vamos dar uma olhada, vou usar alguns dados do conjunto de dados da íris

```python
# Let's first capture a list of the variables we are interested in
import pandas as pd
df=pd.read_csv("assets/iris.csv")
cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

# Now we need to create a grid of subplots of size width and height equal to the number of different variables
# we want to explore, in this case that's a 4x4 grid
fig, axs = plt.subplots(len(cols), len(cols), figsize=(10,10))

# Now we want to iterate across each column in our dataframe and compare it to each other column in our
# DataFrame. We'll do this with a couple of nested for loops
for i in range(len(cols)):
    for j in range(len(cols)):
        # Now we just want to plot a scatter plot comparing the columns i and j. Sometimes this will be the
        # same column, so we would expect to see a diagnol line trend. I'm going to set the marker size to 5
        # just to make things a bit more clear
        axs[i,j].scatter(df[cols[j]], df[cols[i]], s=5)

        # Also, we've seen that when we plot multiple Axes things get cluttered with Axis tickmarks and
        # labels, so let's turn those off
        axs[i,j].get_xaxis().set_visible(False)
        axs[i,j].get_yaxis().set_visible(False)

        # Then we'll turn them back on only if we are the last row...
        if i == len(cols) - 1:
            axs[i,j].get_xaxis().set_visible(True)
            axs[i,j].set_xlabel(cols[j])
            
        # ...and similarly, only show the y-axis labels for the first column.
        if j == 0:
            axs[i,j].get_yaxis().set_visible(True)
            axs[i,j].set_ylabel(cols[i])

# Now let's take a look at our data!
```

Certo, ótimo! Temos um bom exemplo de SPLOM e podemos comparar facilmente o comprimento e a largura das sépalas e dos pedais. e procure tendências rapidamente. Um que salta para mim é que a largura da pétala, a linha inferior, tem uma relação bastante linear com o comprimento da pétala, o terceiro gráfico de dispersão acima. Isso não parece ser verdade se você comparar a largura da sépala com o comprimento da sépala, observando a segunda linha e a célula da primeira coluna.
